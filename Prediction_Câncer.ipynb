```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
dados = pd.read_csv('The_Cancer_data_1500_V2.csv')
```

```python
print(dados.head())
```

```python
print(dados.tail())
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    dados[['Age', 'Gender', 'BMI', 'Smoking', 'PhysicalActivity', 'AlcoholIntake', 'CancerHistory']],
    dados[['Diagnosis']],
    test_size=0.2,
    random_state=49
)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
tree = DecisionTreeRegressor()
```

```python
neural_network = tf.keras.Sequential([
    tf.keras.layers.Dense(89, activation='relu', input_shape=(7,)),
    tf.keras.layers.Dense(46, activation='relu'),
    tf.keras.layers.Dense(15, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
neural_network.summary()
```

```python
neural_network.compile(optimizer='adam', loss='MSE')
```

```python
tree.fit(x_train, y_train)
```

```python
neural_network.fit(x_train, y_train, epochs=489, batch_size=79, verbose=1)
```

```python
test_predictions_tree = tree.predict(x_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions_tree)
MSE = mean_squared_error(y_test, test_predictions_tree)

R2 = r2_score(y_test, test_predictions_tree)

print(f' O erro médio absoluto do algoritmo árvore de decisão é: {MAE}')
print(f' O erro médio quadrático do algoritmo árvore de decisão é: {MSE}')
print(f' O coeficiente de determinação do algoritmo árvore de decisão é: {R2}')
```

```python
test_predictions_neural = neural_network.predict(x_test)
```

```python
MAE_neural = mean_absolute_error(y_test, test_predictions_neural)
MSE_neural = mean_squared_error(y_test, test_predictions_neural)

R2_neural = r2_score(y_test, test_predictions_neural)

print(f' O erro médio absoluto do algoritmo redes neurais é: {MAE_neural}')
print(f' O erro médio quadrático do algoritmo redes neurais é: {MSE_neural}')
print(f' O coeficiente de determinação do algoritmo redes neurais é: {R2_neural}')
```
